# Filling the missing values with another values

#### As, I noticed here in phone column there around 1208 row are empty and this not an easy or you can say it is impossible task for now to get this amount of phone numbers
- So, Here I simply going to replace these values with 'not mentioned(NM)' for short

```python
order_z_data['phone'].fillna(value = 'NM', inplace = True) # inplace is used to change the value at original dataset
```

```python
order_z_data
```

```python
#validating by again checking the phone columns
order_z_data['phone'].isnull().sum()
```

**Again I observed that in location column 21 rows are NaN value** 
- I decided to replace with next row value. as I observe in data most of the location are repeating with certain pattern
- I am bfill() method to do so.

```python
order_z_data['location'].fillna(method = 'bfill', limit = 1, inplace = True)
```

```python
order_z_data
```

```python
# validating by again checking the location column
order_z_data['location'].isnull().sum()
```

**For filling the cuisines column I am taking the same approach to fill as I taken in location column because most of the restaurants are in same location so the chance of same cuisines is more!**

```python
order_z_data['cuisines'].fillna(method = 'bfill', limit = 1, inplace = True)
```

```python
order_z_data
```

```python
# validating by checking again the cuisines column 
order_z_data['cuisines'].isnull().sum()
```

```python
order_z_data['rate'] = order_z_data['rate'].str.replace('/5', ' ')
```

### Here I use ffill to fill the missing data point as for now but we can also use the mean and median to do so.
##### To fill the missing values of rate column I am going use the concept of mean.
- Mean or Median Imputation: In this approach, the missing values in the rating column are replaced with the mean or median value of the available ratings. This method assumes that the missing values are missing at random and that the overall distribution of ratings is relatively stable.

```python
order_z_data['rate'].fillna(method = 'ffill', inplace = True)
```

```python
# validating by checking again the rate column
order_z_data['rate'].isnull().sum()
```
